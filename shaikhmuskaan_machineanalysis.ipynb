```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\maintenance_data\maintenance_data.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
a=['broken','team','provider']
b=['lifetime','pressureInd','moistureInd','temperatureInd']
```

```python
df.head().T
```

```python
#univariate Analysis for categorical variables.
for i in a:
    sns.countplot(x=i,data=df)
    plt.show()
```

```python
sns.countplot(x="team",hue="provider",data=df)
plt.show()
```

```python
sns.countplot(x="broken",hue="provider",data=df)
plt.show()
```

```python
for j in b:
    sns.distplot(df[j],kde=False)
    plt.show()
    print(df[j].describe())
    print(df[j].skew())
```

```python
sns.distplot(df["lifetime"][df["broken"]==1],color='r')
sns.distplot(df["lifetime"][df["broken"]==0],color='b')
plt.show()
```
